```python
# Import Libraries
import pandas as pd
```

```python
# Read in data set
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]
```

```python
# Check the first 5 rows for content
df.head(10)
```

```python
# Change column names for optimal processing
df = df.rename(columns={'Postcode': 'PostalCode', 'Neighbourhood':'Neighborhood'}, inplace=True)
```

```python
df.head(10)
```

```python
df.shape
```

```python
# Remove records not assigned to a Borough
toronto_dropNAB = df[df.Borough != 'Not assigned'].reset_index(drop=True)
```

```python
toronto_dropNAB.head(10)
```

```python
# Combine rows with similar Neighbourhood - separate with a comma 
toronto_groupSN = toronto_dropNAB.groupby(['PostalCode','Borough'], as_index=False).agg(lambda x: ','.join(x))
```
